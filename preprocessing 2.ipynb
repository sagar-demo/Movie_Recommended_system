```python
meta['release_date']=pd.to_datetime(meta['release_date'],errors='coerce')
```

```python
### Getting only 2017 movie as we already have movie up to the year 2016 in preprocessing 1 file
# we don't have enough data for the movie from 2018,2019 and 2020
# we'll deal with it in the upcoming prepocessing files
new_meta=meta.loc[meta.year==2017,['genres','id','title','year']]
```

```python
new_meta
```

```python
new_meta['id']=new_meta['id'].astype(int)
```

```python
## merging the dataset new_meta with credit data
data=pd.merge(new_meta, df ,on='id')
```

```python
data
```

```python
## evaluate an expression node or a string containing a python literal or container display
import ast
data['genres']=data['genres'].map(lambda x : ast.literal_eval(x))
data['cast']=data['cast'].map(lambda x:ast.literal_eval(x))
data['crew']=data['crew'].map(lambda x:ast.literal_eval(x))
```

```python
def make_genreslist(x):
    gen=[]
    st=" "
    for i in x:
        if i.get('name')=='Science Fiction':
            scifi='Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return(st.join(gen))
```

```python
data['genres_list']=data['genres'].map(lambda x:make_genreslist(x))
```

```python
data['genres_list']
```

```python
def get_actor1(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[]:
        return np.NaN
    else:
        return(casts[0])
```

```python
data['actor_1_name']=data['cast'].map(lambda x:get_actor1(x))
```

```python
def get_actor2(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=1:
        return np.NaN
    else:
        return(casts[1])
```

```python
data['actor_2_name']=data['cast'].map(lambda x:get_actor2(x))
```

```python
data['actor_2_name']
```

```python
def get_actor_3(x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name']=data['cast'].map(lambda x:get_actor_3(x))
```

```python
data['actor_3_name']
```

```python
def director(x):
    dt=[]
    st=" "
    for i in x:
        if i.get('job')=='Director':
            dt.append(i.get('name'))
    if dt==[]:
        return np.NaN
    else:
        return(st.join(dt))
```

```python
data['director_name']=data['crew'].map(lambda x:director(x))
```

```python
data['director_name']
```

```python
movie=data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```

```python
movie
```

```python
movie.isnull().sum()
```

```python
movie=movie.dropna(how='any')
```

```python
movie.isnull().sum()
```

```python
movie=movie.rename(columns={'genres_list':'genres'})
movie=movie.rename(columns={'title':'movie_title'})
```

```python
movie['movie_title']=movie['movie_title'].str.lower()
```

```python
movie['comb']=movie['actor_1_name'] + ' ' + movie['actor_2_name']+ ' ' + movie['actor_3_name']+ ' ' + movie['director_name']+ ' ' + movie['genres']
```
