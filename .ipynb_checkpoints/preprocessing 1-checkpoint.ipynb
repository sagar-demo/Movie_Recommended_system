```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
# we have movie only upto 2016
import matplotlib.pyplot as plt
df.title_year.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(20,20))
```

```python
(df.movie_title=='Toy Story\xa0').sum()
```

```python
#in this columns we will check all info about this movie like author name ,actor name
df.loc[df.movie_title=='Toy Story\xa0',['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
## recommendation will be based on these features only
df=df.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
df.head(10)
```

```python
df['actor_1_name']=df['actor_1_name'].replace(np.nan,'unknown')
df['actor_2_name']=df['actor_2_name'].replace(np.nan,'unknown')
df['actor_3_name']=df['actor_3_name'].replace(np.nan,'unknown')
df['director_name']=df['director_name'].replace(np.nan,'unknown')
```

```python
df.head(50)
```

```python
df['genres']=df['genres'].str.replace('|',' ')
```

```python
df.head()
```

```python
## converting movie_title name to lower alphabet
df['move_title']=df['movie_title'].str.lower()
```

```python
df.head()
```

```python
## null terminating char at the end 
df.movie_title[1]
```

```python
## removing the null terminating char at the end
df.movie_title=df.movie_title.apply(lambda x : x[:-1])
```

```python
df.movie_title[1]
```

```python
df.to_csv('data.csv',index=False)
```
